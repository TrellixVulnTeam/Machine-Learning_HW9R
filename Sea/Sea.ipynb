```python
data = pd.read_csv('epa-sea-level.csv')
x = data["Year"]
y= data['CSIRO Adjusted Sea Level']
plt.subplots()
#Draw Scatter
plt.scatter(
    x = x,
    y= y
)
#draw linregress
linre = stats.linregress(x,y)
x_predict = pd.Series([x for x in range(1880,2051)])
#slope: do doc cua duong hoi quy
#intercept: diem chan cua duong hoi quy
y_predict = linre.slope*x_predict + linre.intercept #ax + b
plt.plot(x_predict,y_predict, 'r')
data_new = data[data['Year'] >= 2000]
#draw predict Sea level
new_lingree = stats.linregress(data_new['Year'],data_new['CSIRO Adjusted Sea Level'])
x_new_predict = pd.Series([x for x in range(2000,2051)])
y_new_predict = new_lingree.slope*x_new_predict + new_lingree.intercept
plt.plot(x_new_predict,y_new_predict,'green')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```
