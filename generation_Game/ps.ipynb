```python
data = pd.read_csv('PS.csv')
d = data.groupby(by = ['platforms','genre']).size().reset_index(name = 'total')
sns.set(rc={'figure.figsize':(20,20)})
plt.figure(figsize= (20,20))
sns.catplot(
    data = d,
    x = 'platforms',y = 'total',
    kind = 'bar',hue = 'genre',
    height= 25
)
```
